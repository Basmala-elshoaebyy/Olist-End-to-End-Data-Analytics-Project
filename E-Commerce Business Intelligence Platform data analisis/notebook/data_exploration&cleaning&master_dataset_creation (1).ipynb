```python
print("--- Fixing Delivery Days in master5 (Moving from Truncation to Precision) ---")

# Ensure columns are datetime objects in master5
master5['order_delivered_customer_date'] = pd.to_datetime(master5['order_delivered_customer_date'])
master5['order_purchase_timestamp'] = pd.to_datetime(master5['order_purchase_timestamp'])

# Calculate fractional days for better precision
master5['delivery_days_fixed'] = (master5['order_delivered_customer_date'] - master5['order_purchase_timestamp']) / pd.Timedelta(days=1)

# Round to 1 decimal place
master5['delivery_days_fixed'] = master5['delivery_days_fixed'].round(1)

print("Delivery days updated successfully in master5.")
display(master5[['order_id', 'delivery_days', 'delivery_days_fixed']].head())
```

```python
print("--- Mapping Missing English Categories in master5 ---")
translation_map = {
    'portateis_cozinha_e_preparadores_de_alimentos': 'small_appliances_kitchen_and_food_preparators',
    'pc_gamer': 'pc_gamer'
}

# Apply the fix to master5
master5['product_category_name_english'] = master5['product_category_name_english'].fillna(master5['product_category_name'].map(translation_map))
print("Missing category translations repaired in master5.")
```

```python
print("--- Flagging Anomalies in master5 ---")

# Ensure logistics dates are datetime
master5['order_delivered_carrier_date'] = pd.to_datetime(master5['order_delivered_carrier_date'])
master5['order_approved_at'] = pd.to_datetime(master5['order_approved_at'])

# Create the flag column in master5
master5['is_timeline_anomaly'] = (
    (master5['order_delivered_carrier_date'] < master5['order_purchase_timestamp']) |
    (master5['order_delivered_carrier_date'] < master5['order_approved_at'])
)

print(f"Flagged {master5['is_timeline_anomaly'].sum()} rows with carrier timeline sequencing issues.")
print("You can now filter them using master5[master5['is_timeline_anomaly'] == False]")
```

```python
# Saving the fully cleaned and flagged master5 dataset
master5.to_csv('master5_cleaned.csv', index=False)
print("Cleaning complete. master5_cleaned.csv has been saved.")

# Quick check on the fixes
print(f"Total anomalies flagged: {master5['is_timeline_anomaly'].sum()}")
print(f"Missing translations remaining: {master5['product_category_name_english'].isnull().sum()}")
display(master5[['order_id', 'delivery_days_fixed', 'is_timeline_anomaly']].head())
```
